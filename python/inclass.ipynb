```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
points = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0]
])

plt.figure(figsize=(5, 5))
plt.plot(points[:, 0], points[:, 1],'r-o')
plt.xlabel('x')
plt.ylabel('y')
plt.xlim(-1,1)
plt.ylim(-1,1)
plt.grid(True)
plt.show()
```
