```python
import matplotlib.pyplot as plt
import numpy as np

from matplotlib import cm
```

```python
plt.style.use('_mpl-gallery')
plt.ion()

X = np.arange(-10, 10, 0.75)
Y = np.arange(-5, 23, 0.75)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(X, Y, Z, vmin=Z.min() * 2, cmap=cm.Reds)

ax.set(xticklabels=[],
       yticklabels=[],
       zticklabels=[])

plt.show()
```
